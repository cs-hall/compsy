```python
def rescale(arr):
    """
    Takes an array and returns an array scaled from 0-1 for all values

    Examples
    --------
    >>> test_range = np.arange(-2, 2, 1)
    >>> rescale(test_range)
    array([0.        , 0.33333333, 0.66666667, 1.        ])
    >>> test_linspace = np.linspace(-100, 100, 5)
    >>> rescale(test_linspace)
    array([0.  , 0.25, 0.5 , 0.75, 1.  ])
    """
    return (arr-arr.min()) / (arr.max()-arr.min())

help(rescale)
```

## Defining Defaults 

```python
def rescale(arr, scaled_min=0, scaled_max=1):
    """
    Takes an array and returns an array scaled from the minimum to maximum parameters (by default 0, 1)


    Examples
    --------
    >>> test_range = np.arange(-2, 2, 1)
    >>> rescale(test_range)
    array([0.        , 0.33333333, 0.66666667, 1.        ])
    >>> test_linspace = np.linspace(-100, 100, 5)
    >>> rescale(test_linspace)
    array([0.  , 0.25, 0.5 , 0.75, 1.  ])
    """
    
    assert scaled_min < scaled_max, "min is not smaller than max"

    result = (arr-arr.min()) / (arr.max()-arr.min()) 
    if not ((scaled_min==0) & (scaled_max==1)):
        result = (result * (scaled_max-scaled_min)) + scaled_min
    
    return result
```

```python
print(rescale(test_linspace, scaled_min=-1, scaled_max=10))
print(rescale(test_range, scaled_min=-10, scaled_max=-1))
```

## Readable Code 

```python
def fence(center_text, wrapper):
    """
    This adds a wrapper text to both sides of the center_text. Note: it attempts to make the inputs a string, if unsuccesful it will throw an error

    Examples
    --------    
    >>> fence("word", "*")
    '*word*'
    >>> fence(2, 4)
    '424'
    >>> fence(0, "hello")
    'hello0hello'
    """
    # Coerce the stuff to strings
    wrapper = str(wrapper)
    center_text = str(center_text)

    return wrapper+center_text+wrapper
```

## Testing Assertions

```python
def get_total(values):
    assert len(values) > 0
    for element in values:
        assert int(element)
    values = [int(element) for element in values]
    total = sum(values)
    assert total > 0
    return total
```

```python
# tests that each value can be an integer
get_total([1,"a"])
```

```python
patients = [[70, 1.8], [80, 1.9], [150, 1.7]]

def calculate_bmi(weight, height):
    return weight / (height ** 2)

for patient in patients:
    weight, height = patients[0]
    bmi = calculate_bmi(height, weight)
    print(f"Patient's BMI is: {bmi} (height={height}, weight={weight}")
```

```python
# be explicit in the function call
patients = [[70, 1.8], [80, 1.9], [150, 1.7]]

def calculate_bmi(weight, height):
    return weight / (height ** 2)

for patient in patients:
    weight, height = patient
    bmi = calculate_bmi(height=height, weight=weight)
    print(f"Patient's BMI is: {bmi} (height={height}, weight={weight}")
```
